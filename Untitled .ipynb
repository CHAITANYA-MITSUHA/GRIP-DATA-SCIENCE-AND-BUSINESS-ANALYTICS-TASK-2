# GRAPH TO FIND K

```python
x = dataset.iloc[:, [1,2,3]].values
inertias = []

for i in range(1, 8):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x)
    inertias.append(kmeans.inertia_)
    
plt.plot(range(1, 8), inertias)
plt.title("Elbow Graph")
plt.xlabel('number of clusters')
plt.ylabel('inertia')
plt.show()
```
